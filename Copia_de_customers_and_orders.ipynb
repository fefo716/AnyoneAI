```python
# open and evaluate orders.csv
df1 = pd.read_csv('https://raw.githubusercontent.com/anyoneai/notebooks/main/customers_and_orders/data/orders.csv')
print(df1.info())
#1
print('Answer one:')
print(len(list(df1['OrderID'].unique())))
print(' \n ')
#2
print('Answer two:')
oocc = df1.groupby('OrderID',axis=0).count()
print(oocc['CustomerID'].mean())
print(' \n ')
#3
print('Answer three:')
print(oocc.nlargest(1,'CustomerID'))
print(' \n ')
#4
print('Answer four:')
df1['year'] = pd.DatetimeIndex(df1['Date']).year
df1['month'] = pd.DatetimeIndex(df1['Date']).month
print(df1.head())
date = df1.groupby(['month','year'],axis=0).count()
print(date.loc[10,2021])
print(' \n ')
#5
print('Answer five:')
customer = df1.groupby(['year','CustomerID']).sum()
print(customer.loc[2021])
#print(customer.loc[2021].sort_values('OrderTotal',ascending=False))
print(customer.loc[2021].sort_values('Price',ascending=False))
print(' \n ')
#6
print('Answer six:')
best_month = df1.groupby('month').sum()
print(best_month.sort_values('Price',ascending=False))
```
